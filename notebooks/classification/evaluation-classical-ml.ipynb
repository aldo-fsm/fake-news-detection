```python
def evaluateAll(Xtest, ytrue, paths):
  results = []
  for path in predictorPaths:
    print(path)
    predictor = joblib.load(path)
    ypred = evaluation.evaluate(predictor, Xtest, ytrue)
    results.append(dict(
        model=path,
        accuracy=accuracy_score(ytrue, ypred)
    ))
    print('\n')
  return results
```
